This code check the distribution of verteces in all the files we have

```python
from os import listdir
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

path_data = '/home/leonardo/LCP_B_WORKDIR/PROJECT_knots/data_extracted/'
files = sorted(listdir(path_data))
files_classified = [s for s in files if s.endswith('ooo.dat')]

classes = Counter({})

for p in files_classified:
    
    with open(path_data + p, mode='r') as f:
        
        data = np.array(f.readlines())
        N_position = list(range(1,len(data)+1,4))
        
        data = data[N_position]
        
        for idx,d in enumerate(data):
            data[idx] = d.strip('\n')
        
        classes = classes + Counter(data)  
            
```

```python
# plot
k, n = [], []
cutoff = 2000
for ver, number in classes.most_common():
    if number > cutoff:
        k.append(ver)
        n.append(number)
        
plt.figure(figsize=(22,6))
plt.ylabel('Number of protein with that len', fontsize=14)
plt.xticks(rotation=90)
plt.bar(k,n)
# plt.hlines(1000, k[0], k[-1], 'r')
```
